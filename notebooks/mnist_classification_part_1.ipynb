### Dimensionality reduction

As the first step, MNIST dataset will be reduced to 2 dimension using three different techniques (PCA, Isomap, t-SNE) and results will be compared by plotting the reduced datasets.

```python
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap, TSNE

digit_colors = ["red","yellow","black","blue","plum","magenta","aqua","lime","orange","olive"]
```

```python
import time
target_dimensions = 2
#pca
start = time.time()
pca = PCA(n_components=target_dimensions)
pca.fit(x_train)
x_train_pca = pca.transform(x_train)

end = time.time()
print("PCA duration: {:.2f} s".format(end - start))
```

```python
#isomap
start = time.time()
iso = Isomap(n_neighbors=3, n_components=target_dimensions)
# Since the whole training dataset is too big for Isomap (60000 samples), only half of the dataset will be used
iso.fit(x_train[0:30000,:])
x_train_iso = iso.transform(x_train)

end = time.time()
print("Isomap duration: {:.2f} s".format(end - start))
```

```python
#t-SNE
start = time.time()
t_sne = TSNE(n_components=target_dimensions)
x_train_tsne = t_sne.fit_transform(x_train)

end = time.time()
print("t-SNE duration: {:.2f} s".format(end - start))
```

```python
from matplotlib import colors
fig, ax = plt.subplots()
fig.set_size_inches(14,8)
for color in digit_colors:
    index = digit_colors.index(color)
    mask = y_train == index
    x = x_train_pca[mask,0]
    y = x_train_pca[mask,1]
    ax.scatter(x,y,c=color,alpha=0.5,s=8,label=index)

ax.legend(markerscale=4)
ax.set_title("PCA")
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14,8)
for color in digit_colors:
    index = digit_colors.index(color)
    mask = y_train == index
    x = x_train_iso[mask,0]
    y = x_train_iso[mask,1]
    ax.scatter(x,y,c=color,alpha=0.5,s=8,label=index)

ax.legend(markerscale=4)
ax.set_title("Isomap")
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14,8)
for color in digit_colors:
    index = digit_colors.index(color)
    mask = y_train[0:len(x_train_tsne)] == index
    x = x_train_tsne[mask,0]
    y = x_train_tsne[mask,1]
    ax.scatter(x,y,c=color,alpha=0.5,s=8,label=index)

ax.legend(markerscale=4)
ax.set_title("t-SNE")
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(14,8)
for color in digit_colors:
    index = digit_colors.index(color)
    mask = y_train == index
    x = x_train_iso[mask,0]
    y = x_train_iso[mask,1]
    ax.scatter(x,y,c=color,alpha=0.5,s=8,label=index)

ax.legend(markerscale=4)
ax.set_title("Isomap (30 000 samples used for Isomap training)")
```

### Comparison of dimensions size and execution time for PCA

```python
from xgboost import XGBClassifier
import time

dimension_sizes = [2,5,10,40]
for dimensions in dimension_sizes:
    print(30*"=")
    #pca
    pca = PCA(n_components=dimensions)
    pca.fit(x_train)
    x_train_pca = pca.transform(x_train)

    #classification
    start = time.time()
    print("Training set shape: {}".format(x_train_pca.shape))
    model = XGBClassifier()
    model.fit(x_train_pca, y_train)
    end = time.time()
    print("Training duration: {:.2f} s".format(end - start))

    #testing
    x_test_pca = pca.transform(x_test)
    y_pred = model.predict(x_test_pca)
    predictions = [round(value) for value in y_pred]
    accuracy = metrics.accuracy_score(y_test, predictions)
    print("Accuracy: {:.2f}%".format(accuracy * 100.0))
```

### Data preparation and saving for model tunning

The training set will be transformed to shape 60 000 x 40, as discussed above. Test set will have shape 10 000 x 40.

```python
import os
data_directory = "data"
if not os.path.exists(data_directory):
    os.makedirs(data_directory)
    
#labels saving, so that original dataset with labels doesn't need to be loaded during model tuning
np.savetxt(data_directory+"/y_train.csv", y_train, delimiter=",", fmt="%d")
np.savetxt(data_directory+"/y_test.csv", y_test, delimiter=",", fmt="%d")
```

```python
FINAL_DIMENSIONS = 40
```

#### PCA

```python
pca = PCA(n_components=FINAL_DIMENSIONS)
pca.fit(x_train)
x_train_pca_final = pca.transform(x_train)
x_test_pca_final = pca.transform(x_test)
np.savetxt(data_directory + "/x_train_pca_final.csv", x_train_pca_final, delimiter=",", fmt='%f')
np.savetxt(data_directory + "/x_test_pca_final.csv", x_test_pca_final, delimiter=",", fmt='%f')
```

#### Isomap

```python
iso = Isomap(n_neighbors=3, n_components=FINAL_DIMENSIONS)
# Since the whole training dataset is too big for Isomap (60 000 samples), only half of the dataset will be used
iso.fit(x_train[0:30000,:])
x_train_iso_final = iso.transform(x_train)
x_test_iso_final = iso.transform(x_test)
np.savetxt(data_directory + "/x_train_isomap_final.csv", x_train_iso_final, delimiter=",", fmt='%f')
np.savetxt(data_directory + "/x_test_isomap_final.csv", x_test_iso_final, delimiter=",", fmt='%f')
```
